```python
import staintools
%matplotlib inline
```

# i1

```python
i1 = staintools.read_image('./data/i1.png')
staintools.plot_image(i1, fig_size=4, title="Original Image")

mask = staintools.get_luminosity_mask(i1)
staintools.plot_image(mask, fig_size=4, title="Tissue Mask")
```

```python
standardizer = staintools.BrightnessStandardizer()
i1_transformed = standardizer.transform(i1)
staintools.plot_image(i1_transformed, fig_size=4, title="Brightness Standardized Image")

mask = staintools.get_luminosity_mask(i1)
staintools.plot_image(mask, fig_size=4, title="Tissue Mask")
```
